```python
#Se importan las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Función para leer los datos de posición en el eje x

```python
def position_data(csv_data):
    # Cargar los datos desde el archivo CSV
    data = pd.read_csv(csv_data)
    
    # Filtrar las columnas que empiezan con "X_"
    x_columns = [col for col in data.columns if col.startswith('X_')]
    
    # Concatenar todas las columnas de posiciones X en un solo array
    x_positions = pd.concat([data[col] for col in x_columns], ignore_index=True)
    
    return x_positions
    
```

```python
#Ahora definimos los arreglos conteniendo las posiciones en el eje x para cada una de las simulaciones
datos_4_discos = position_data('4disk_data.csv')
datos_muchos_discos = position_data('many_disks.csv')
```

### Función para graficar el histograma:

```python
def histograma_gen(positions,fig_name):
    plt.hist(positions,bins=30,color='blue',alpha=0.7)
    plt.title('Histograma combinado de posiciones X')
    plt.xlabel('Posición X')
    plt.ylabel('Frecuencia')
    plt.savefig(f'{fig_name}.png', dpi=300)
    plt.show()
```

## Histograma para la simulación con cuatro discos:

```python
histograma_gen(datos_4_discos,'Histograma_4_discos')
```

```python
histograma_gen(datos_muchos_discos,'Histograma_Muchos_discos')
```
